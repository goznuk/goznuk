```python
import numpy as np
def water(height):
  maxheight=max(height)
  height=np.array(height)
  water=0
  for i in range(1,maxheight+1):
    k = height[np.min(np.where(height>=i)):np.max(np.where(height>=i))+1]
    tf = k>=i
    water+=len(tf)-sum(tf)
  return water
```

```python
water([0,1,0,2,1,0,1,3,2,1,2,1])
```

```python
water([0,2,0,1,2,0,0,1,3,0])
```

```python
water([0,3,2,4,1,0,3,0,4,0])
```
